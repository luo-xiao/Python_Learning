```python
from numpy import linspace, cos, pi, ceil, floor, arange
from matplotlib.pylab import plot, show, axis
```

```python
#sample a signal limited to 30Hz with a sampling rate of 600Hz
f = 30
tmin = -0.2
tmax = 0.2
t = linspace(tmin,tmax,300)  #start, end, num_of_samples
x = cos(2*pi*t) + cos(2*pi*f*t)
plot(t,x)
```

```python
#sample a signal limited to 30Hz with a sampling rate of 600Hz
f = 30
tmin = -0.2
tmax = 0.2
t = linspace(tmin,tmax,300)
x = cos(2*pi*t) + cos(2*pi*f*t)
plot(t,x)

#sample the signal with a sampling rate of 60Hz, using Nyquist rate
T = 1/60.0
nmin = ceil(tmin/T)
nmax = floor(tmax/T)
n = arange(nmin, nmax)
print(n)
x1 = cos(2*pi*n*T) + cos(2*pi*f*n*T)
plot(n*T, x1, 'bo')
```

```python
#sample a signal limited to 30Hz with a sampling rate of 600Hz
f = 30
tmin = -0.2
tmax = 0.2
t = linspace(tmin,tmax,300)
x = cos(2*pi*t) + cos(2*pi*f*t)
plot(t,x)

#sample the signal with a sampling rate of 60Hz, using Nyquist rate
T = 1/60.0
nmin = ceil(tmin/T)
nmax = floor(tmax/T)
n = arange(nmin, nmax)  # number of samples
x1 = cos(2*pi*n*T) + cos(2*pi*f*n*T)
plot(n*T, x1, 'bo')

#sample the signal with a sampling rate of 25Hz which is under the Nyquist rate
T=1/25.0
nmin = ceil(tmin/T)
nmax = floor(tmax/T)
n = arange(nmin,nmax)
print(n)
x2 = cos(2*pi*n*T) + cos(2*pi*f*n*T)
plot(n*T, x2, '-r', markersize=6)
#axis([-0.2, 0.2, 1.8, 2.6])
show()
```
